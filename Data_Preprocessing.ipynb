# Loading Dependencies

```python
import pandas as pd
import numpy as np

```

# Data Loading

```python
df = pd.read_csv('filtered_data_JourneyGenie.csv')
df.head()
```

```python
df.info()
```

```python
cols = [' FaxNumber', ' PinCode', 'countyCode', ' Description', ' cityCode', ' HotelCode']
df = df.drop(cols, axis=1)
```

```python
df.info()
```

```python
star_mapping = {
    'All': 5,
    'FourStar': 4,
    'ThreeStar': 3,
    'TwoStar': 2,
    'onestar' : 1,
    'FiveStar' : 5
}

df[' HotelRating'] = df[' HotelRating'].map(star_mapping)
```

```python
df.head()
```

# Data Cleaning

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Data Transformation

```python
no_duplicates = df.drop_duplicates()
```

```python
no_duplicates.info()
```

```python
import re 
mixed_column = ' cityName'


def contains_numerics(text):
    return bool(re.search(r'\d', str(text)))

numeric_mask = df[mixed_column].apply(contains_numerics)

numerical_count = numeric_mask.sum()

df = df[~numeric_mask]
```

```python
df.info()
```

```python
df.head()
```

```python
def count_attractions(attractions):
    # Split attractions by the starting pattern, if found
    split_attractions = attractions.split('<br /> <p>')
    if len(split_attractions) > 1:
        # Remove starting line and count attractions
        attractions_list = split_attractions[1].split('<br /> ')
        return len(attractions_list)
    else:
        return 0  # If pattern not found, return 0 attractions

# Apply the function to each row
df['NumOfAttractions'] = df[' Attractions'].apply(count_attractions)
```
